```python
data = pd.read_csv('reviews_courses.csv')

# Konwersja kolumny Timestamp na format daty
data['Timestamp'] = pd.to_datetime(data['Timestamp'])

# Grupowanie danych według daty i obliczanie średniej oceny dla każdego dnia
daily_average_rating = data.groupby(data['Timestamp'].dt.date)['Rating'].mean()

# Tworzenie wykresu
plt.figure(figsize=(10, 6))
daily_average_rating.plot()
plt.title('Średnia ocena w zależności od daty')
plt.xlabel('Data')
plt.ylabel('Średnia ocena')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Wczytanie danych
data = pd.read_csv('reviews_courses.csv')

# Konwersja kolumny Timestamp na format daty
data['Timestamp'] = pd.to_datetime(data['Timestamp'])

# Grupowanie danych według daty i obliczanie średniej oceny dla każdego miesiąca
monthly_average_rating = data.groupby(data['Timestamp'].dt.to_period('M'))['Rating'].mean()

# Tworzenie wykresu
plt.figure(figsize=(10, 6))
monthly_average_rating.plot()
plt.title('Średnia ocena w zależności od miesiąca')
plt.xlabel('Miesiąc')
plt.ylabel('Średnia ocena')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
