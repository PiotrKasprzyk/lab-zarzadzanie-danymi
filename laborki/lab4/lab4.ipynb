```python
import pandas as pd
import matplotlib.pyplot as plt
import pytz
```

```python
# Wczytanie danych
df = pd.read_csv('reviews_courses.csv')

# Podstawowa analiza danych
print(df.head())  # Wyświetlenie pierwszych kilku wierszy

# Histogram z rozkładem wartości dla przykładowej kolumny (Rating)
df['Rating'].hist()
plt.title('Histogram of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

# Średnia wartość dla kolumny Rating
average_rating = df['Rating'].mean()
print(f'Srednia wartość Rating: {average_rating}')
```

```python
# Konwersja kolumny Timestamp na datetime
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Filtrowanie danych z zakresu od 1 lipca 2020 do 31 grudnia 2020
start_date = '2020-07-01'
end_date = '2020-12-31'
filtered_df_date = df[(df['Timestamp'] >= start_date) & (df['Timestamp'] <= end_date)]
print(filtered_df_date.head())
```

```python
# Konwersja kolumny Timestamp na datetime
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Filtrowanie danych z zakresu od 1 lipca 2020 do 31 grudnia 2020
start_date = '2020-07-01'
end_date = '2020-12-31'
filtered_df_date = df[(df['Timestamp'] >= start_date) & (df['Timestamp'] <= end_date)]
print(filtered_df_date.head())
```

```python
# Średnia ocen dla wszystkich kursów
average_rating_all_courses = df['Rating'].mean()
print(f'Srednia ocen dla wszystkich kursów: {average_rating_all_courses}')

# Średnia ocen dla kursu "The Python Course: From Beginner to Expert" w okresie od 1 lipca 2020 do 31 grudnia 2020
filtered_course = df[(df['Course Name'] == 'The Python Course: From Beginner to Expert') & 
                     (df['Timestamp'] >= start_date) & 
                     (df['Timestamp'] <= end_date)]
average_rating_specific_course = filtered_course['Rating'].mean()
print(f'Srednia ocen dla kursu "The Python Course: From Beginner to Expert" w wybranym okresie: {average_rating_specific_course}')

# Liczba wierszy bez komentarzy
no_comment_rows = df[df['Comment'].isna()].shape[0]
print(f'Liczba wierszy bez komentarzy: {no_comment_rows}')

# Liczba wierszy z komentarzami zawierającymi wybrane słowo (accent)
comment_with_word = df[df['Comment'].str.contains('accent', na=False)].shape[0]
print(f'Liczba wierszy z komentarzami zawierającymi słowo "accent": {comment_with_word}')
```

```python
# Wykres z podziałem ocen wg dni
df['Date'] = df['Timestamp'].dt.date
daily_average_rating = df.groupby('Date')['Rating'].mean()

daily_average_rating.plot(kind='bar', figsize=(10, 5))
plt.title('Średnia ocena wg dni')
plt.xlabel('Data')
plt.ylabel('Średnia ocena')
plt.show()
```
