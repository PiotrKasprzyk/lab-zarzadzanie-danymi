```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# Wczytanie danych
df = pd.read_csv('reviews_courses.csv')



```

```python

df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Grupowanie danych wg dni
daily_avg = df.groupby(df['Timestamp'].dt.date)['Rating'].mean()

# Wykres z rozkładem ocen wg dni
daily_avg.plot(kind='bar', figsize=(12, 6))
plt.title('Średnia ocena wg dni')
plt.xlabel('Data')
plt.ylabel('Średnia ocena')
plt.show()
```

```python
# Grupowanie danych wg tygodni
weekly_avg = df.groupby(df['Timestamp'].dt.to_period('W'))['Rating'].mean()

# Wykres z rozkładem ocen wg tygodni
weekly_avg.plot(kind='bar', figsize=(12, 6))
plt.title('Średnia ocena wg tygodni')
plt.xlabel('Tydzień')
plt.ylabel('Średnia ocena')
plt.show()
```

```python
# Grupowanie danych wg miesięcy
monthly_avg = df.groupby(df['Timestamp'].dt.to_period('M'))['Rating'].mean()

# Wykres z rozkładem ocen wg miesięcy
monthly_avg.plot(kind='bar', figsize=(12, 6))
plt.title('Średnia ocena wg miesięcy')
plt.xlabel('Miesiąc')
plt.ylabel('Średnia ocena')
plt.show()
```

```python
# Grupowanie danych wg miesięcy i kursów
monthly_course_avg = df.groupby([df['Timestamp'].dt.to_period('M'), 'Course Name'])['Rating'].mean().unstack()

# Wykres z rozkładem ocen wg miesięcy, oddzielnie dla każdego kursu
monthly_course_avg.plot(kind='bar', figsize=(15, 8))
plt.title('Średnia ocena wg miesięcy (oddzielnie dla każdego kursu)')
plt.xlabel('Miesiąc')
plt.ylabel('Średnia ocena')
plt.legend(title='Nazwa kursu')
plt.show()

# Wykres sumy ocen dla poszczególnych kursów wg miesięcy
monthly_course_sum = df.groupby([df['Timestamp'].dt.to_period('M'), 'Course Name'])['Rating'].sum().unstack()
monthly_course_sum.plot(kind='bar', stacked=True, figsize=(15, 8))
plt.title('Suma ocen dla poszczególnych kursów wg miesięcy')
plt.xlabel('Miesiąc')
plt.ylabel('Suma ocen')
plt.legend(title='Nazwa kursu')
plt.show()
```

```python
# Dodanie kolumny z dniem tygodnia
df['DayOfWeek'] = df['Timestamp'].dt.day_name()

# Grupowanie danych wg dni tygodnia
day_of_week_avg = df.groupby('DayOfWeek')['Rating'].mean()

# Wykres zadowolenia użytkowników wg dni tygodnia
day_of_week_avg.plot(kind='bar', figsize=(12, 6))
plt.title('Średnia ocena wg dni tygodnia')
plt.xlabel('Dzień tygodnia')
plt.ylabel('Średnia ocena')
plt.show()
```
